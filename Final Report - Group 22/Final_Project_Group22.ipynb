# 1. Introduction

Road traffic accounts for about 20% of total emissions in the Netherlands. Electric cars present an effective solution to reduce these emissions. An electric car does not produce emissions, such as nitrogen dioxide. In the 2019 Climate Agreement ("Het Klimaatakkoord"), the Dutch government committed to reducing CO₂ emissions by 49% by 2030 and by 95% by 2050. Transitioning to clean transportation is essential to achieving these targets, and by 2040, all new passenger cars in the Netherlands are mandated to be fully electric, running on power from batteries, hydrogen fuel cells, or solar panels (Rijksoverheid, n.d.).

As of early 2023, there were over 511,000 plug-in cars in the Netherlands, representing roughly 6% of all passenger cars. Of these, about 65% were fully electric, meaning that only around 4% of all passenger cars were entirely emission-free (CBS, 2023). This is far from the government's targets for 2030 and 2050. To meet these goals, significant changes are needed, particularly in the infrastructure supporting electric vehicles. One key factor influencing the adoption of electric vehicles is the availability of charging points. Without sufficient and accessible charging infrastructure, the transition to 100% electric vehicles will be challenging, especially outside urban areas where access to charging points is more limited. This leads to the research question:

**What is the impact of the availability of electric charging points on the usage of electric vehicles in the Netherlands?**

Charging infrastructure varies widely by region, affecting people's willingness to switch to electric vehicles. If drivers cannot easily charge their EVs, they are less likely to adopt them. By identifying these regional differences, policymakers can improve the charging network for electric vehicles. This would help facilitate the transition to electric vehicles and ensure that clean transportation is accessible to everyone.

To address this research question, the following sub-questions will guide our analysis:

1. How does the availability of charging infrastructure differ between urban and rural areas in the Netherlands?
    - What is the number of charging points per municipality in the Netherlands?
    - How is this number divided between urban and non-urban areas?
2. What role does the income level of a municipality’s residents play in the ratio of electric vehicles to charging points?
3. How has the development of charging infrastructure influenced the growth of registered electric vehicles over the years, and vice versa? 

The initial two questions will be answered with data from 1 January 2021, as this date is supported by detailed information from a variety of sources. This particular point of reference guarantees consistency across our datasets, thus enabling us to effectively examine the distribution of charging points and electric vehicles and draw meaningful conclusions. For the final sub-question, a broader timeframe (2014–2023 for national data and 2019–2023 for municipal data) is utilized due to limited municipal-level data prior to 2019.

This report is structured according to these sub-questions. The initial objective is to investigate the differences in the accessibility of charging infrastructure between urban and rural regions within the Netherlands. Next, the impact of municipal inhabitants' income levels on the ratio of electric vehicles to charging points will be examined. After that, the relationship between the expansion of charging infrastructure and the growth of EV registrations over time is explored. Finally, the conclusions are presented and discussed in order to answer our research question. 

# 2. Difference in Charging Infrastructure between Urban and Rural Areas

This chapter examines the distribution of charging infrastructure in urban and rural areas, with a focus on the number of charging points in each municipality and the division of these points between urban and non-urban settings. By understanding these regional variations, insights can be gained into the infrastructure gaps that may hinder equitable EV adoption, and future investments in charging infrastructure can be identified as a priority.

The first map shows the amount of addresses per km2 for each municipality in the Netherlands. The darker the colour, the higher the address density of the municipality. The areas with the highest densities are mostly found in the Randstad, which is a densely populated metropolitan region in the west side of the Netherlands that includes cities like Amsterdam, Rotterdam and The Hague. In contrast, rural regions, especially in the north and east side of the Netherlands have much lower address densities. This map helps to visualise urban and rural addresses per square kilometre, which is useful to understand where there may be a higher demand for electrical charging points. 

This map categories each municipality by its urbanisation level, which is categorised by the CBS (Central Bureau of Statistics in the Netherlands) which uses address density in order to categorise provinces into five different urbanisation levels: 
- very highly urban: average amount of 2500 or more addresses per km2;
- highly urban: average number of 1500 to 2500 addresses per km2;
- moderately urban: average number of 1000 to 1500 addresses per km2;
- little urban: average number of 500 to 1000 addresses per km2;
- non-urban: on average less than 500 addresses per km2. 
The randstad region again stands out as highly urbanised, this indicates that a greater need for charging infrastructure is needed in comparison to the east side of the Netherlands. This map is essential for understanding how urbanisation impacts the infrastructure distribution, especially comparing the amount of the availability of charging points in urban versus rural areas. 



This map displays the number of charging points available in each municipality. The darker the colour, the more charging points there are in the municipality. Urban municipalities in the randstad show a higher amount of charging points whereas rural areas show significantly fewer. It is striking to see that cities in Noord-Brabant such as Tilburg, Breda and Eindhoven also have a high amount of charging points. On the other hand, rural areas typically have less than 1000 charging points. This map showcases the imbalance of charging infrastructure, as urban areas have more charging points. 

This map categorises the density of municipalities by the density of charging points per municipality. This map helps to identify areas with more accessible charging points and those with limited availability. It clearly shows a saturation of charging points in urban areas and a scarcity in rural areas.

This scatterplot examines the relationship between the address density and the number of charging points per municipality. Each point represents a municipality, categorized by its urbanisation level. The trend line shows a positive correlation, indicating that municipalities with higher densities in addresses generally have more charging points. This plot thereby confirms the trend observed in the previous maps. 

Urban areas, especially within the Randstad, have significantly more charging points due to their higher population density. Rural areas have far fever charging points, which can discourage the use of electric vehicles in these regions. The amount of charging points varies widely per municipality, highly urbanised municipalities often have more than 5000 charging points, with some municipalities exceeding 10.000 charging points. In rural municipalities the number is typically below 1000. This displays the higher demand for charging points in urban areas. 

*How does the availability of charging infrastructure differ between urban and rural areas in the Netherlands?*

In summary, urban areas have a much denser network of charging points, especially in highly urban municipalities to meet higher demand. While rural areas have much less charging points. There is a clear urban-rural divide in charging points availability, with urban areas well-equipped and rural areas not as much. Expanding the overall charging infrastructure will be essential to better support electric vehicle charging across the country.

# 4. Influence of Charging Point Infrastructure on Registered Electric Vehicles

A strong correlation between the number of charging points and the number of electric vehicles seems inevitable, as electric power is essential for operating electric cars. To better understand this relationship, the saturation of charging points has been examined by dividing the number of electric vehicles by the number of charging points (Klimaatmonitor, n.d.). It is important to note that private charging points have not been included in this analysis due to insufficient registration data. This approach provides a clear indication of any surplus or shortage of charging points, offering a more precise insight into the balance between vehicles and charging capacity.

This graph above shows the relationship between the growth of battery-electric vehicles (BEVs) and the saturation of charging points from 2014 to 2023. It indicates that a larger growth in BEVs is associated with increased pressure on charging points each year. As the number of BEVs rises, the ratio of BEVs per charging point increases, logically raising the strain on the existing infrastructure. A higher saturation correlates with a reduced relationship to growth one and two years later, with correlation coefficients of 0.93 for growth one year later and 0.84 for growth two years later, compared to 0.97 for the current year. This suggests that the increased pressure on charging points may discourage people from purchasing electric vehicles in the future. The saturation of charging points in relation to growth one or two years later is shown in the following two plots.


Furthermore, we examined the impact of charging point saturation on the increase in the number of charging points. A correlation of 0.84 was observed in the same year, which increases to 0.94 for the growth in the following one and two years. Therefore, the growth in charging points, along with a decrease in the growth rate of electric vehicles, should create a more favorable situation over the years, provided that the reduction in electric vehicle growth is significant enough in relation to the increase in charging points.


Unfortunately, the situation is quite the opposite, as noticeable in the graph above. The decrease in the growth rate of BEVs and the increase in the number of charging points do not offset each other sufficiently to reduce saturation. By 2021, the saturation level has reached 4.5, meaning there are 4.5 electric vehicles per charging point. While the growth in electric vehicles has stabilized in recent years, the interplay between the number of charging points and electric vehicles remains insufficient to create a healthier balance in the charging infrastructure.

Subsequently, the saturation of charging points per municipality has been examined over the years. In the absence of data, the plots cover the period from 2019 to 2023. The overall color of the charts has become somewhat lighter over time, indicating an improvement in saturation. However, this is not a universal phenomenon, with some municipalities exhibiting a contrasting trend.

Over the years, certain high-saturation areas have demonstrated only marginal improvement, suggesting that the expansion of charging infrastructure has not kept pace with the rate of adoption of battery electric vehicles (BEVs). This trend highlights the necessity for a more rapid adaptation of infrastructure in these areas. One illustrative case is that of the municipality of Haarlemmermeer. Haarlemmermeer is a municipality situated in close proximity to Amsterdam and is characterized by an urbanized population. The graph below illustrates the growth in the number of BEVs per resident in relation to the level of saturation. The rate of growth in the number of BEVs per resident is high. In 2019, one in ten residents had a BEV. By 2023, the ratio had risen to one in three residents. However, the rate of saturation was unable to keep pace with the accelerated growth. In 2022, saturation levels were more favorable, and in the subsequent year, the expansion of BEVs was notable. It is plausible that saturation levels may have contributed to this phenomenon. Nevertheless, there are numerous potential explanations for the significant increase in BEVs per resident, and there is no evidence to suggest that saturation can fully account for the observed growth.


Nevertheless, yearly changes demonstrate that certain regions have effectively diminished saturation through strategic infrastructure investments, which could inform the formulation of expansion strategies in other regions. Addressing these trends and regional needs will be essential for supporting the transition to electric vehicles. To illustrate, the municipality of Hellendoorn has demonstrated an improvement in saturation levels since 2021. While the growth in BEVs per resident is not substantial, it is nevertheless evident. The expansion of charging stations has contributed to a notable increase in saturation levels. However, it should be noted that Hellendoorn is a municipality in the eastern region of the Netherlands, a rural area with a relatively small population. The total number of residents is also considerably smaller than in, for example, Haarlemmermeer, which may facilitate the municipality's ability to adapt to the growth in BEVs.


The two municipalities highlighted serve merely as illustrative examples of the saturation observed from 2019 to 2023. In both cases and in the remainder of the Netherlands, the number of BEVs and charging points increased. Nevertheless, it is not possible to draw any conclusions regarding the impact of the charging point infrastructure on the growth of BEVs, or vice versa. A multitude of additional factors may exert an influence on saturation. For example, it is possible that the number of electric cars on the road may be influenced by the decision of companies to provide electric cars exclusively to their employees. Or, alternatively, an increase in the number of BEVs may be attributed to subsidies. Furthermore, the number of charging points is not solely dependent on the number of electric vehicles. In this study, only (semi-)public charging points are considered. It is possible that individuals may construct private charging points in certain areas. For a municipality, there are numerous considerations beyond the mere construction of charging points. The prosperity of a municipality is also a factor that must be taken into account. It is conceivable that a municipality may have desired to construct additional charging points, but was unable to do so.

*How has the development of charging infrastructure influenced the growth of registered electric vehicles over the years, and vice versa?*

In conclusion, while a correlation definitely exists, it remains uncertain whether charging points directly influence the growth of BEVs or if the reverse is true. Further research is needed to clarify this relationship.


# 5. Conclusion and Discussion

The transition to electric vehicles (EVs) in the Netherlands is a crucial step toward reducing road traffic emissions and achieving national climate goals. Our analysis explored several key factors that impact EV adoption, particularly the availability and distribution of charging infrastructure across urban and rural areas, the influence of municipal income levels, and the relationship between infrastructure development and EV growth.

Our findings highlight a significant disparity in charging infrastructure availability between urban and rural areas. Urban regions, especially densely populated municipalities within the Randstad, benefit from a higher density of charging points, making EV usage more feasible. In contrast, rural areas generally lack sufficient infrastructure, posing a barrier to widespread EV adoption outside metropolitan centers. Addressing this urban-rural divide is essential to support an equitable shift to electric mobility.

Income levels in municipalities also play a role, as higher-income areas tend to have a more favorable ratio of EVs to charging points. This suggests that socioeconomic factors may influence both EV adoption and infrastructure placement, underscoring the importance of targeted policies to ensure that lower-income municipalities are not left behind in the EV transition.

Finally, our examination of infrastructure growth and EV adoption revealed a positive but complex relationship. While increased infrastructure correlates with higher EV registration, a lag in infrastructure expansion relative to EV growth has led to high saturation levels in some regions, which may discourage future adoption. This finding emphasizes the need for a proactive approach in expanding charging infrastructure to meet current and projected demand.

In summary, while the Netherlands has made substantial progress toward developing an EV-supportive infrastructure, challenges remain in achieving uniform access and sufficient capacity across all regions. To facilitate the nationwide transition to electric vehicles, policymakers should prioritize expanding infrastructure in rural and underserved areas, consider socioeconomic factors in infrastructure planning, and accelerate efforts to close the gap between infrastructure growth and EV adoption.
